# Cleaning dataset: Air Quality Database

Source: WHO Air Quality Database 2023, version 6.0
Local filename: WHO_annual_mean_pm25.csv

Concentrations of fine particulate matter (PM2.5) of all countries from 2011 - 2019
Mean values and range[min, max]

```python
import pandas as pd
```

```python
df = pd.read_csv('WHO_,annual_mean_pm25.csv')
df.head()
```

### Make a new table with only total values

```python
df_total = df
df_total = df_total.iloc[:, [0, 1, 6, 11, 16, 21, 26, 31, 36, 41, 46]]
df_total
```

```python
# Rename column names

new_columns = ['Country','2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010' ]
df_total.columns = new_columns

```

```python
# Reorder columns to start from 2010-2019

new_column_order = ['Country', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
df_total = df_total[new_column_order]
```

```python
# Drop top two rows

df_total = df_total.drop(df.index[[0,1]])

df_total
```

### Add new column 'Continent' for easy grouping

```python
# Use country_converter to assign a Continent to a Country as a new column 
# pip install country_converter

import country_converter as coco

cc = coco.CountryConverter()
df_total['continent'] = df_total['Country'].apply(lambda x: cc.convert(x, to='continent'))
df_total.insert(loc=1, column='Continent', value=df_total['continent'])
df_total = df_total.drop(columns=['continent'])

df_total
```

### Use values only (mean)

- remove range [min - max] for now
- can be used later

```python
# New df_mean with mean values only

df_mean = df_total.copy()

columns_to_process = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

for col in columns_to_process:
    df_mean[col] = df_mean[col].str.extract(r'^([\d.]+)')

df_mean
```

```python
# New df_min with min values only

df_min = df_total.copy()

def extract_min(value):
    start = value.index('[') + 1
    end = value.index('-')
    min_val = float(value[start:end])
    return min_val

columns_to_process = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

for col in columns_to_process:
    df_min[col] = df_min[col].apply(lambda x: extract_min(x))

df_min
```

```python
# New df_mean with max values only

df_max = df_total.copy()

def extract_max(value):
    start = value.index('-')+1 
    end = value.index(']')
    max_val = float(value[start:end])
    return max_val

columns_to_process = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

for col in columns_to_process:
    df_max[col] = df_max[col].apply(lambda x: extract_max(x))

df_max
```

```python
# Export to .csv

df_mean.to_csv('pm25_mean.csv', index=False)
df_min.to_csv('pm25_min.csv', index=False)
df_max.to_csv('pm25_max.csv', index=False)
```

### Safe levels of pm2.5 = 10

Find countries which are above and below these safe levels


```python
# Below safe levels in 2019

countries_below_10 = df_mean[df_mean['2019'].astype(float) < 10]

countries_below_10
```

```python
grouped_counts = df_mean['2019'].astype(float).apply(lambda x: 'Above 10' if x > 10 else 'Below 10').value_counts()

grouped_counts
```
